```python
paises = pd.read_csv("countries.csv")
```

```python
paises
```

```python
Ecuador = paises[paises.country == "Ecuador"]
```

```python
Honduras = paises[paises.country == "Honduras"]
```

```python
plt.plot(Ecuador.year, Ecuador.population / Ecuador.population.iloc[0]*100)
plt.plot(Honduras.year, Honduras.population / Honduras.population.iloc[0]*100)
plt.legend(["Hodnuras", "Ecuador"])
plt.title("Poblacion Honduras y Ecuador")
plt.xlabel("Ao")
plt.ylabel("Crecimiento de la poblacion (1er ao 100)")
plt.show()
```
